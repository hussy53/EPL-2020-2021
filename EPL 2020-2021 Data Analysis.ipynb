```python
# Import the libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Load Dataset

epl_df = pd.read_csv('C:\\Users\\Administrator\\Desktop\\EPL_20_21.csv', encoding= 'unicode_escape')
epl_df.head()
```

```python
epl_df.info()
```

```python
epl_df.describe() # Works on numeric columns only
```

```python
epl_df.isna().sum()
```

```python
# Create 2 new columns

epl_df['MinsPerMatch'] = (epl_df['Mins'] / epl_df['Matches']).astype(int)
epl_df['GoalsPerMatch'] = (epl_df['Goals'] / epl_df['Matches']).astype(float)
epl_df.head()
```

```python
# Total Goals

Total_Goals = epl_df['Goals'].sum()
print(Total_Goals)
```

```python
# Penalty Goals

Total_Penalties = epl_df['Penalty_Goals'].sum()
print(Total_Penalties)
```

```python
# Penalty Attempts

Total_Attempted_Penalties = epl_df['Penalty_Attempted'].sum()
print(Total_Attempted_Penalties)
```

```python
# Pie chart for penalties scored vs missed

plt.figure(figsize=(13, 6))
pl_not_scored = Total_Attempted_Penalties - Total_Penalties
data = [pl_not_scored, Total_Penalties]
labels = ['Penalties missed', 'Penalties scored']
color = sns.color_palette("mako")
plt.pie(data, labels=labels, colors=color, autopct='%.0f%%')
```

```python
# Unique Positions

epl_df['Position'].unique()
```

```python
epl_df[epl_df['Position'] == 'FW']
```

```python
# Players from different nations, np for non-numeric

np.size((epl_df['Nationality'].unique()))
```

```python
# Most players from which countries

nationality = epl_df.groupby('Nationality').size().sort_values(ascending=False)
nationality.head(10).plot(kind='bar', figsize=(12, 6), color=sns.color_palette('magma'))
```

```python
# Clubs with maximum players in squad

epl_df['Club'].value_counts().nlargest(5).plot(kind='bar', color=sns.color_palette('viridis'))
```

```python
# Clubs with least players in squad

epl_df['Club'].value_counts().nsmallest(5).plot(kind='bar', color=sns.color_palette('viridis'))
```

```python
# Players based on age group

under20 = epl_df[epl_df['Age'] <= 20]
age20_25 = epl_df[(epl_df['Age'] > 20) & (epl_df['Age'] <= 25)]
age25_30 = epl_df[(epl_df['Age'] > 25) & (epl_df['Age'] <= 30)]
above30 = epl_df[epl_df['Age'] > 30]
```

```python
x = np.array([under20['Name'].count(), age20_25['Name'].count(), age25_30['Name'].count(), above30['Name'].count()])
mylabels = ["Under 20", "Between 20 & 25", "Between 25 & 30", "Above 30"]
plt.title('Total players with Age Groups', fontsize=20)
plt.pie(x, labels=mylabels, autopct='%.1f%%')
plt.show()
```

```python
# Total players under 20 in each club

under_20 = epl_df[epl_df['Age'] < 20]
under_20['Club'].value_counts().plot(kind='bar', color=sns.color_palette('cubehelix'))
```

```python
# Players under 20 in ManU

under_20[under_20['Club'] == 'Manchester United']
```

```python
# Players under 20 in Chelsea

under_20[under_20['Club'] == 'Chelsea']
```

```python
# Average age of players in each club

plt.figure(figsize=(12, 6))
sns.boxplot(x='Club', y='Age', data=epl_df)
plt.xticks(rotation=90)
```

```python
num_players = epl_df.groupby('Club').size()
data = (epl_df.groupby('Club')['Age'].sum()) / num_players
data.sort_values(ascending=False)
```

```python
# Total assists from each club

assists_by_clubs = pd.DataFrame(epl_df.groupby('Club', as_index=False)['Assists'].sum())
sns.set(style='whitegrid', color_codes=True)
ax = sns.barplot(x='Club', y='Assists', data=assists_by_clubs.sort_values(by='Assists'), palette='Set2')
ax.set_xlabel('Club', fontsize=30)
ax.set_ylabel('Assists', fontsize=20)
plt.xticks(rotation=75)
plt.rcParams['figure.figsize'] = (20, 8)
plt.title('Plot of Clubs vs Total Assists')
```

```python
# Top 10 assists

top_assists = epl_df[['Name', 'Club', 'Assists', 'Matches']].nlargest(n=10, columns='Assists')
top_assists
```

```python
# Goals from each club

goals_by_clubs = pd.DataFrame(epl_df.groupby('Club', as_index=False)['Goals'].sum())
sns.set(style='whitegrid', color_codes=True)
ax = sns.barplot(x='Club', y='Goals', data=goals_by_clubs.sort_values(by='Goals'), palette='rocket')
ax.set_xlabel('Club', fontsize=30)
ax.set_ylabel('Goals', fontsize=20)
plt.xticks(rotation=75)
plt.rcParams['figure.figsize'] = (20, 8)
plt.title('Plot of Clubs vs Total Assists', fontsize=20)
```

```python
# Most goals by players

top_goals = epl_df[['Name', 'Club', 'Goals', 'Matches']].nlargest(n=10, columns='Goals')
top_goals
```

```python
# Goals per match

goals_per_match = epl_df[['Name', 'GoalsPerMatch', 'Matches', 'Goals']].nlargest(n=10, columns='GoalsPerMatch')
goals_per_match
```

```python
# Pie Chart - Goals with assist and without assists

plt.figure(figsize=(14, 7))
assists = epl_df['Assists'].sum()
data = [Total_Goals - assists, assists]
labels = ['Goalswithout assists', 'Goals with assists']
color = sns.color_palette('Set1')
plt.pie(data, labels=labels, colors=color, autopct='%.0f%%')
plt.show()
```

```python
# Top 10 players with most yellow cards

epl_yellow = epl_df.sort_values(by='Yellow_Cards', ascending=False)[:10]
plt.figure(figsize=(20, 6))
plt.title('Players with the most yellow cards')
c = sns.barplot(x=epl_yellow['Name'], y=epl_yellow['Yellow_Cards'], label='Players', color='Yellow')
plt.ylabel('Number of yellow cards')
#c.set_xticklabels(c.get_xticklabels(), rotation=45)
c
```
